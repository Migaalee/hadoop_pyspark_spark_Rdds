# Open CSV with Pandas


```python
# Load the Pandas libraries with alias 'pd' 
import pandas as pd 
# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 
data = pd.read_csv("Taxi_Trips_151MB.csv",sep= ';', header = None)
data.columns = ['Trip id','Taxi ID','Trip Start Timestamp',
                'Trip End Timestamp','Trip Seconds','Trip Miles','Pickup RegionID','Dropoff RegionID',
                'Pickup Community Area','Dropoff Community Area','Fare','Tips',
                'Tolls','Extras','Trip Total','Payment Type','Company',
                'Pickup Centroid Latitude','Pickup Centroid Longitude','Pickup Centroid Location','Dropoff Centroid Latitude','Dropoff Centroid Longitude','Dropoff Centroid Location']

# Preview the first 5 lines of the loaded data 
data
```

```python
%%time 
from __future__ import print_function
import pyspark
from operator import add as sum 

sc = pyspark.SparkContext('local[*]')

try :
    data = sc.textFile("Taxi_Trips_151MB.csv")
    non_empty_lines = data.filter( lambda line : len(line) > 0 ) # Remove empty lines 
    date = non_empty_lines.map( lambda line : line.split(';')[2]) # Gives the date
    month = date.map(lambda line : line[0:2]).filter(lambda line : line != '') # Picks just the month number
    month_tuples = month.map(lambda month : (month, 1)) # create a tuple
    occurrences = month_tuples.reduceByKey(sum) # calculate the number of occurrences for each unique tuple
    sort = occurrences.sortByKey(ascending=True) # order the dates by ascending order
    for (k,v) in sort.collect():
        print(k, v )
    sc.stop()
except Exception as e:
    print(e)
    sc.stop()
```

```python
%%time 
from __future__ import print_function
import pyspark
from operator import add as sum 

sc = pyspark.SparkContext('local[*]')

try :
    data = sc.textFile("Taxi_Trips_151MB.csv")
    lines =  data.flatMap(lambda line : line.split(";;; "))
    keyValue = lines.map(lambda line : (line.split(';')[6],[line.split(';')[7]]))
    filt = keyValue.filter(lambda line : line[0] != '' and line[1][0] != '')
    set_conv = filt.map(lambda kv : (kv[0],set(kv[1]))) # convert value to set
    union = set_conv.reduceByKey(lambda p,q : p | q)
    for (k,v) in union.collect():  
        print(k,v)
    sc.stop()
except Exception as e:
    print(e)
    sc.stop()
```

# Exercise 3
- *What is the expected charge/cost of a taxi ride, given the pickup region ID, the weekday (0=Monday, 6=Sunday) and time in format “hour AM/PM”?*\
- Expected output:
<img src="Exercise3.png" alt="Alt text that describes the graphic" title="Title text" />




```python
%%time 
from __future__ import print_function
import pyspark
from operator import add as sum 
import datetime 

def findDay(date): 
    return  datetime.datetime.strptime(date, '%m/%d/%Y').weekday() # Convert a date with the format 
# month/day/year to a number that is equivalent to a week day (0=Monday,6=Sunday)

sc = pyspark.SparkContext('local[*]') 

try :
    data = sc.textFile("Taxi_Trips_151MB.csv")
    lines =  data.flatMap(lambda line : line.split(";;; "))
    keyValue = lines.map( lambda line : line.split(';')).filter(lambda line : line[6] != '' and line[14] != '') \
               .map(lambda line :(line[6] , line[2].split(),line[14].replace(',', '')))
    desiredFormat = keyValue.map(lambda line : (line[0] + "_" + str(findDay(line[1][0])) + "_" 
                                          + line[1][1][0:2] + line[1][2],float(line[2]))) # convert data to desired format
    avg = desiredFormat.mapValues(lambda v: (v, 1)).reduceByKey(lambda a,b: (a[0]+b[0], a[1]+b[1])) \
                  .mapValues(lambda v: round(v[0]/v[1],2)) #compute the average
    
    for (k,v) in  avg.collect():  
        print(k,v)
    sc.stop()
except Exception as e:
    print(e)
    sc.stop()
```
